```python
list_of_simulation_wait_times = []
wait_times_SVO = []
thetas_SVO = []
traffic_simulator = ts.TrafficSimulator()
traffic_simulator.batch = True
traffic_simulator.initialize_map(10,10,4,0.1)
traffic_simulator.coordinator = tc.BatchTrafficCoordinator()

#############IMPORTANT PARAMATERS
p_left = 0.3
p_right = 0.3
p_human = 0.20
traffic_simulator.coordinator.bid_tax = 0.5
number_iterations_before_plotting = 10
################################


time_between_cars_multiplier = 0.1
dt = 0.1
next_car = -1
number_iterations_between_cleaning_time_matrix = 4

wait_times_humans = []
thetas_SVO_humans = []
wait_times_AVs = []
```

## FCFS-ish, Reserving Entire Intersection

```python
for i in range(100):
    if (i % number_iterations_between_cleaning_time_matrix) == 0:
        traffic_simulator.coordinator.cleanup_matrix_interval(traffic_simulator.time)

        
    ################ Generate a new car
    if traffic_simulator.iterations_until_next_car <= 0:
        new_car = traffic_simulator.generate_car(p_human, p_left,p_right)
        if new_car: #new_car = None if collision
            new_car.svo_theta = rand.choice([0,np.pi/4.0,np.pi/2.0])
            traffic_simulator.add_robot(new_car)
            time_between_cars = new_car.time_required_intersection(AV.TurnDirection.STRAIGHT,new_car._maxSpeed)
            number_iterations = time_between_cars_multiplier*time_between_cars/traffic_simulator.deltaT
            posson_lambda = 1/number_iterations        
            traffic_simulator.iterations_until_next_car = rand.expovariate(posson_lambda)
    else:
        traffic_simulator.iterations_until_next_car -= 1

    
    traffic_simulator.incrementTime(dt)
    if (i % number_iterations_before_plotting) == 0:
        traffic_simulator.plotRobots(fsize=(4,4),plot_lanes_flag=True,annotate_flag=True,plot_intersection_region_flag=True,plot_reservation_patches_flag=False,plot_only_intersection_flag=False)    
#         plt.savefig(results_directory+'/z'+str(i).zfill(4)+'.png')
#         plt.close()
        plt.show()
    print(i,end="\r")
control_times = [(car.time_entered_intersection-car.time_entered_control) for car in traffic_simulator.totalRobotArray if car.time_entered_intersection!=-1]

wait_times_humans += [(car.time_entered_intersection-car.time_entered_control) for car in traffic_simulator.totalRobotArray if car.time_entered_intersection!=-1 and car.human]
thetas_SVO_humans += [(car.svo_theta) for car in traffic_simulator.totalRobotArray if car.time_entered_intersection!=-1 and car.human]
wait_times_AVs += [(car.time_entered_intersection-car.time_entered_control) for car in traffic_simulator.totalRobotArray if car.time_entered_intersection!=-1 and not car.human]
# print(i)

list_of_simulation_wait_times += [control_times]
```

```python
print([robot.agentID for robot in sorted(traffic_simulator.active_robots,key=lambda r: r.agentID)])

self = traffic_simulator.coordinator
```

```python
sorted_queue = sorted(self.bid_queue,key=lambda b: b[0])
final_reservations = []
#LET'S Assume that first thing in queue will have noone in front
for request in sorted_queue:
    bid, current_car, intersection_lane = request

    number_retries = 0
    no_reservation = True
    replan_increment_sec = 0.2
    if current_car in self.car_ids_in_queue: no_reservation = False


    incoming_lane = current_car.current_lane
    if len(self.lane_queues[incoming_lane])>0:
        (previous_car, previous_car_start_time) = self.lane_queues[incoming_lane][-1]
        distance_to_previous_car = current_car.pose.distance(previous_car.pose)
        car_intersection_start_time = previous_car_start_time + distance_to_previous_car/current_car._maxSpeed
    else: # No preceeding cars (i.e. start time = time to interesection)
        car_intersection_start_time = current_car.simTime + current_car.get_time_to_intersection()

    while no_reservation and number_retries <= 10:                
        # 1:  Add offset to the estimated time at intersection
        car_intersection_start_time += number_retries*replan_increment_sec

        #2:  Compute the list of positions the vehicle will traverse
        list_of_positions_time = self.forward_simulate(car_intersection_start_time,intersection_lane,current_car)

        #3:  For each pose, check that the body of the car doesn't conflict with an existing reservation
        ## Add (t_start,t_end),i,j to list of potential reservation tuples
        searching_for_conflicts = True
        list_of_reservation_tuples = []
        while searching_for_conflicts:
            for (x,y,phi,t) in list_of_positions_time:
                points_covered_by_car = self.get_points_covered_by_car(x,y,phi,current_car.car_width,current_car.car_length)
                for point in points_covered_by_car:
                    (i,j) = self.matrix_index_from_xy(point.x,point.y)
                    if self.conflicts_in_interval_bool((t-self.time_buffer_interval,t+self.time_buffer_interval,agentID),i,j):
                        conflicting_trajectory = True #TODO: Check this breaks
                        break       
                    else:
                        list_of_reservation_tuples += [((t-self.time_buffer_interval,t+self.time_buffer_interval,agentID),i,j)]
            searching_for_conflicts = False

        #4: Now add the trajectory to the reservation matrix
        if not conflicting_trajectory:
            for reservation_tuple in list_of_reservation_tuples:
                (t_start,t_end,agentID),i,j = reservation_tuple
                if (0 <= i <= self.m-1) and (0 <= j <= self.n-1):      
                    self.add_interval((t_start,t_end),i,j,agentID)                                        
                    self.time_available_matrix[i,j] = t_end+self.time_buffer_interval #I'm adding a fixed buffer size
            n_tiles_reserved = len(list_of_reservation_tuples)
            no_reservation = False
        number_retries += 1

    # Add reservation information
    final_reservations += [(car_intersection_start_time,t_end)]
    self.lane_queues[current_car.current_lane] += [(current_car,car_intersection_start_time)]
    self.car_ids_in_queue.add(current_car.agentID)
self.bid_queue = []   
for lane in self.lane_queues.keys():
    for (car,start_time) in self.lane_queues[lane]:
        car.reservation = (start_time, 100000)
self.lane_queues = {"1": [], "-1": [], "2": [], "-2": []}
```
